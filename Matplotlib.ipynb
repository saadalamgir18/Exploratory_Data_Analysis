```python
a = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(a,)
plt.show()
```

```python
a = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(a, vert = False, patch_artist = False)
plt.show()
```

```python
a = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(a, vert = False, patch_artist = True)
plt.show()
```

```python
a = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(a, vert = True, patch_artist = True)
plt.show()
```

```python
alist = ['Python','JAVA','C++', 'Node']
values = [215, 130, 245, 110]
colors = ['green', 'red', 'yellow', 'blue']
explode = (0.1, 0, 0, 0)
plt.pie(values,  labels = alist, colors = colors, explode = explode, autopct = '%1.1f%%')
plt.plot()
```
