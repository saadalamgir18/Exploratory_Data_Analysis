# Distribution Plots

<li>distplot
<li>joinplot
<li>pairplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
df[['total_bill', 'tip']]
```

# JoinPlot

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind= 'hex')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind= 'reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

# DistPlot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde = False, )
```

# Categorical Plots

<li> boxplt
<li>violinplot
<li>countplot
<li>barplot

# Countplot

```python
sns.countplot(data = df,x = 'sex')
```

```python
sns.countplot(data = df,x = 'smoker')
```

```python
sns.countplot(data = df,x = 'day')
```

```python
sns.countplot(data = df,y = 'day')
```

# Barplot

```python
sns.barplot(x= 'sex', y = 'total_bill', data = df)
```

```python
sns.barplot(x= 'sex', y = 'tip', data = df)
```

# Boxplot

```python
sns.boxplot(x = 'sex', y = 'total_bill', data = df)
```

```python
sns.boxplot(data=df,)
```

```python
sns.boxplot(y= 'day', x = 'total_bill', hue = 'smoker', data = df)
```

# Violanplot

```python
sns.violinplot(x = 'total_bill', y = 'day', data = df)
```
